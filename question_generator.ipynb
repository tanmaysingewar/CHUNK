```python
import logging
import sys
import pandas as pd

import nest_asyncio
nest_asyncio.apply()

logging.basicConfig(stream=sys.stdout, level=logging.INFO)
logging.getLogger().addHandler(logging.StreamHandler(stream=sys.stdout))
```

```python
from llama_index.core import SimpleDirectoryReader, VectorStoreIndex, Response
from llama_index.core.evaluation import DatasetGenerator
reader = SimpleDirectoryReader("./research_papers")
documents = reader.load_data()
data_generator = DatasetGenerator.from_documents(documents)
```

```python
questions = data_generator.generate_questions_from_nodes()
questions
```

```python
# Get the length of questions 
len(questions)
```

```python
import random
random_questions = random.sample(questions, 100)
```

```python
import csv
# Specify the file name
file_name = 'csv/research_papers-questions.csv'

# Write the list of questions to a CSV file
with open(file_name, mode='w', newline='',encoding='utf-8') as file:
    writer = csv.writer(file)
    
    # Write the header
    writer.writerow(["Question"])
    
    # Write each question to a new row
    for question in random_questions:
        writer.writerow([question])

print(f"Questions have been written to {file_name}")
```
